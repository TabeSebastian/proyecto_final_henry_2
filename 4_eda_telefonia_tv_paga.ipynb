```python
#importo librerias
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# abro los datasets a utilizar
Penetración_total_televisión = pd.read_csv('datasests/telefonia_tv_paga/Penetración_total_televisión.csv')


Penetración_total_televisión.head()
```

```python
# Crear un gráfico de líneas para mostrar la evolución de los accesos
fig, ax = plt.subplots(figsize=(10, 6))

# Gráfico de líneas para Accesos TV por suscripcion por cada 100 habitantes
ax.plot(Penetración_total_televisión['Año'].astype(str) + '-Q' + Penetración_total_televisión['Trimestre'].astype(str), Penetración_total_televisión['Accesos TV por suscripcion por cada 100 hab'], marker='o', label='TV por Suscripción por 100 hab')

# Gráfico de líneas para Accesos TV satelital por cada 100 habitantes
#x.plot(Penetración_total_televisión['Año'].astype(str) + '-Q' + Penetración_total_televisión['Trimestre'].astype(str), Penetración_total_televisión['Accesos TV satelital por cada 100 hab'], marker='o', label='TV Satelital por 100 hab')

# Gráfico de líneas para Accesos TV por suscripcion por cada 100 hogares
ax.plot(Penetración_total_televisión['Año'].astype(str) + '-Q' + Penetración_total_televisión['Trimestre'].astype(str), Penetración_total_televisión['Accesos TV por suscripcion por cada 100 hogares'], marker='o', label='TV por Suscripción por 100 hogares')

# Gráfico de líneas para Accesos TV satelital por cada 100 hogares
#ax.plot(Penetración_total_televisión['Año'].astype(str) + '-Q' + Penetración_total_televisión['Trimestre'].astype(str), Penetración_total_televisión['Accesos TV satelital por cada 100 hogares'], marker='o', label='TV Satelital por 100 hogares')

# Etiquetas y título
ax.set_xlabel('Año y Trimestre')
ax.set_ylabel('Accesos por cada 100')
ax.set_title('Evolución de Accesos de TV por Suscripción y Satelital')

# Mostrar leyenda
ax.legend()

# Rotar las etiquetas del eje x para una mejor visualización
plt.xticks(rotation=45)

# Mostrar el gráfico
plt.show()
```

Conclusión:

El propósito de este análisis fue evaluar el estado actual del sector de la television y comprender su evolución a lo largo de los años y trimestres. El gráfico resultante revela de manera evidente que a medida que avanzan los años, la demanda de tv tiende a disminuir (2022 a la izquierda y 2014 a la derecha).

Este fenómeno plantea varias consideraciones significativas. En primer lugar, indica que el mercado de la tv es, en efecto, un mercado decreciente. En este contexto, realizar inversiones sustanciales en este sector podría no ser una estrategia especialmente rentable a largo plazo.

La disminución de la demanda de tv puede atribuirse a diversos factores. En particular, la proliferación de las plataformas de streaming" o "servicios de streaming". Estas plataformas ofrecen contenido audiovisual, como películas, series, documentales y programas de televisión, que los usuarios pueden ver en línea (streaming) a través de Internet en lugar de descargarlo en sus dispositivos. Las plataformas de streaming se han vuelto muy populares en los últimos años, brindando a los espectadores la flexibilidad de ver contenido en sus propios horarios a través de múltiples dispositivos, como televisores, computadoras, tabletas y teléfonos móviles.

En resumen, los datos analizados reflejan un mercado en declive, lo que sugiere que las empresas deben considerar cuidadosamente sus estrategias de inversión y marketing. La tv está cediendo terreno a las tecnologías más modernas y móviles, y comprender esta dinámica es esencial para adaptarse a las cambiantes necesidades y preferencias de los consumidores.
