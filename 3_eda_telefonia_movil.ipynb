```python
#importo librerias a usar

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# abro los datasets a utilizar
hogares_comerciales_gobierno_telefonía_fija = pd.read_csv('datasests/telefonia_movil/hogares_comerciales_gobierno_telefonía fija.csv')

### inicio limpieza
# Realiza la operación de renombrar la columna 'Año'
hogares_comerciales_gobierno_telefonía_fija = hogares_comerciales_gobierno_telefonía_fija.rename(columns={'AÃ±o': 'año'})
# Filtra los datos por año igual a 2022 y trimestre igual a 4
datos_ultimo_anio = hogares_comerciales_gobierno_telefonía_fija[(hogares_comerciales_gobierno_telefonía_fija['año'] == 2022) & (hogares_comerciales_gobierno_telefonía_fija['Trimestre'] == 4)]

datos_ultimo_anio.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Seleccionar las columnas a graficar
columnas = ['Accesos telefonia fija (hogares)',
            'Accesos telefonia fija (comercial)', 'Accesos telefonia fija (gobierno)',
            'Accesos telefonia fija (otros)']

# Etiquetas personalizadas para el eje x
etiquetas = ['Hogares', 'Comercial', 'Gobierno', 'Otros']

# Obtener la lista de provincias únicas
provincias_unicas = datos_ultimo_anio['Provincia'].unique()

# Definir la cantidad de gráficos por fila
graficos_por_fila = 4
num_filas = (len(provincias_unicas) - 1) // graficos_por_fila + 1

# Crear un diseño de subgráficos
fig, axs = plt.subplots(num_filas, graficos_por_fila, figsize=(16, num_filas * 4))

# Crear gráficos de barras para cada provincia
for i, provincia in enumerate(provincias_unicas):
    datos_provincia = datos_ultimo_anio[datos_ultimo_anio['Provincia'] == provincia]
    
    # Sumar los datos para cada columna
    totals = datos_provincia[columnas].sum()
    
    # Calcular las coordenadas del subgráfico
    fila = i // graficos_por_fila
    columna = i % graficos_por_fila

    # Crear un gráfico de barras en el subgráfico correspondiente
    axs[fila, columna].bar(etiquetas, totals, color='skyblue')
    axs[fila, columna].set_title(f'telefonia en {provincia} (2022)')
    axs[fila, columna].set_ylabel('Cantidad')

# Ajustar el diseño y mostrar los gráficos
plt.tight_layout()
plt.show()
```
