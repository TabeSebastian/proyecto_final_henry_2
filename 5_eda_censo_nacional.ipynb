```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# abro los datasets a utilizar
accesos_internet_fijo = pd.read_csv('datasests/censo_nacional/acceso_internet_fijo.csv', encoding='latin1', delimiter=';')

### limpieza
accesos_internet_anio = accesos_internet_fijo[accesos_internet_fijo['mes']== 'Diciembre']

accesos_internet_anio.head()
```

```python
periodos = accesos_internet_anio['Período']
residencial = accesos_internet_anio['Residencial']
organizaciones = accesos_internet_anio['Organizaciones']

plt.stackplot(periodos, [residencial, organizaciones], labels=['Residencial', 'Organizaciones'], alpha=0.5)

plt.xlabel('Período')
plt.ylabel('Total')
plt.title('Total por Período (Residencial y Organizaciones)')
plt.legend()
plt.show()
```

conclusion :

Con el objetivo de profundizar en el análisis del sector de Internet, me propuse obtener información del Censo Nacional Argentino. A través de este proceso, logré acceder a datos que muestran la evolución de la adopción de Internet, diferenciando entre hogares residenciales y organizaciones.
