```python
import requests
import pandas as pd
import csv
from io import StringIO


# URL del archivo CSV
autorizacion_apy = 'V5eMFqG6QhUowcbG10HoHEYmUORRoOfG9HdjZ9n5'
api_localidades_conectividad_internet = 'LISTA-DE-LOCAL-CON-CONEC/data.csv'
url = f"http://api.datosabiertos.enacom.gob.ar/api/v2/datastreams/{api_localidades_conectividad_internet}/?auth_key={autorizacion_apy}"

# Descargar el archivo CSV
response = requests.get(url)

# Verificar si hubo errores en la descarga
response.raise_for_status()  

# Leer el contenido de la respuesta como texto
content = response.text

# Crear un objeto StringIO para leer el contenido como archivo
csv_file = StringIO(content)

# Leer el archivo CSV con pandas
df = pd.read_csv(csv_file, encoding='latin-1')

# Guarda el DataFrame en el archivo CSV
df.to_csv('E:/Mis documentos/henry/proyecto final 222/carpeta demo/datasests/localidades_conectividad_internet.csv', index=False)
```
